## 10.1 Email Creator
Create a program that reads a file and creates a series of emails.

### Console:
```powershell

================================================================
To:	art@venere.org
From:	noreply@deals.com 
Subject: Deals!

Hi Art,

We've got some great deals for you. Check our website!
```

### Specifications:
- Use `email.csv` file:
- Use `email_template.txt` file:
- When the program starts, it should read the email addresses and first names from the file, merge them into the mass email template, and display the results on the console.
- All email addresses should be converted to **lowercase**.
- All first names should be converted to **title case**.
- If you add names to the list of email addresses, the program should create more emails.
- If you modify the template, the program should change the content of the email that’s created. 



````python
import csv

def main():
    print("Email Creator\n")

    # Get emails from csv file
    emails = readCSV('email.csv')

    # Convert email addresses to lower case and firstnames to title casee
    for email in emails:
        email[2] = email[2].lower()
        email[0] = email[0].title()

    # Get email template
    emailTemplate = readTXT('email_template.txt')

    # Replace {email} and {firstname}, then print to console for each email
    for email in emails:
        print("================================================================")
        modifiedEmailTemplate = emailTemplate.replace("{email}", email[2])
        modifiedEmailTemplate = modifiedEmailTemplate.replace("{first_name}", email[0])
        print(modifiedEmailTemplate)
        

def readCSV(filename):
    emails = []
    with open (filename, 'r') as file:
        reader = csv.reader(file)
        for line in reader:
            emails.append(line)
    return emails


def readTXT(filename):
    with open (filename, 'r') as file:
        templateString = file.read()
    return templateString
        
main()
````
